Sentiment Analysis

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
result = classifier("The Hugging Face transformers library is fantastic!")
print(result)
# Output: [{'label': 'POSITIVE', 'score': 0.9998...}]
```

Text Genertation

```python
from transformers import pipeline

text_generator = pipeline("text-generation", model="Qwen/Qwen2.5-1.5B")
prompt = "The secret to baking a really good cake is"
generated_text = text_generator(prompt, max_length=50, num_return_sequences=1)
print(generated_text[0]['generated_text'])
```

machine Translation

```python
from transformers import pipeline

translator = pipeline("translation_en_to_de", model="Helsinki-NLP/opus-mt-en-de")
result = translator("The process of using transformers in Python is straightforward.")
print(result)
# Output: [{'translation_text': 'Der Prozess der Verwendung von Transformatoren in Python ist einfach.'}]
```

Chat

```python
import torch
from transformers import pipeline

chat = [
    {"role": "system", "content": "You are a sassy, wise-cracking robot as imagined by Hollywood circa 1986."},
    {"role": "user", "content": "Hey, can you tell me any fun things to do in New York?"}
]

pipeline = pipeline(task="text-generation", model="meta-llama/Meta-Llama-3-8B-Instruct", dtype=torch.bfloat16, device_map="auto")
response = pipeline(chat, max_new_tokens=512)
print(response[0]["generated_text"][-1]["content"])
```
